```python
import numpy as np
import pandas as pd
import warnings 
warnings.filterwarnings('ignore')
```

```python
#load dataset
df = pd.read_csv('Hospital_Readmissions_Reduction_Program.csv')
```

```python
#footnote has a lot of missing data , and therefore we will drop it.
```

```python
df.drop('Footnote',axis=1 ,inplace= True)
```

```python
#converting some non-numeric index data to numeric.
```

```python
df['Number of Discharges'] = pd.to_numeric(df['Number of Discharges'],errors='coerce').fillna(0).astype(int)
df['Excess Readmission Ratio'] = pd.to_numeric(df['Excess Readmission Ratio'],errors='coerce').fillna(0).astype(float)
df['Predicted Readmission Rate'] = pd.to_numeric(df['Predicted Readmission Rate'],errors='coerce').fillna(0).astype(float)
df['Expected Readmission Rate'] = pd.to_numeric(df['Expected Readmission Rate'],errors='coerce').fillna(0).astype(float)
df['Number of Readmissions'] = pd.to_numeric(df['Number of Readmissions'],errors='coerce').fillna(0).astype(int)
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df.head()
```

filling the actual zeros with median instead of mode to maintain data consistency 

```python
for col in ['Number of Discharges','Excess Readmission Ratio','Predicted Readmission Rate',
            'Expected Readmission Rate','Number of Readmissions']:
    median_value = df[col].median()
    df[col] = df[col].replace(0,median_value)
```
