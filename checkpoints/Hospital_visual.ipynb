```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('cleaned_hospital_Readmissions.csv')
```

```python
df.head()
```

```python
sns.pairplot(data=df)
```

```python
#renaming the Measure name for better data understanding 

df['Measure Name']=df["Measure Name"].replace({'READM-30-HF-HRRP':'Heart_Failure','READM-30-HIP-KNEE-HRRP':'Hip_Knee_Surgery',
'READM-30-PN-HRRP':'Pneumonia','READM-30-AMI-HRRP':'Acute_Myocardial_Infarction', 'READM-30-COPD-HRRP':'Chronic_Obstructive_Pulmonary'})
```

```python
plt.figure(figsize=(10,8))
sns.boxenplot(data = df , x = "Measure Name", y = "Number of Discharges",)
plt.tight_layout()
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(data=df,x ="Measure Name",y = "Excess Readmission Ratio")
plt.tight_layout()
```

```python
plt.figure(figsize=(10,8))
sns.barplot(data=df,x ='Measure Name', y = "Expected Readmission Rate")
plt.tight_layout()
```

```python
plt.figure(figsize=(10,8))
plt.title("Readmission Rate")
sns.boxplot(data=df,x = "Measure Name", y= 'Predicted Readmission Rate')
plt.tight_layout()
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(data = df, x = "Measure Name", y = "Number of Readmissions")
plt.tight_layout()
```

```python
sns.displot(data=df,x = "Number of Readmissions",bins = 50)
```

```python
sns.displot(data=df,x = "Number of Discharges",bins = 60)
```

```python
sns.jointplot(data = df,x = 'Number of Discharges',y = 'Number of Readmissions')
```

```python
cor = df[["Number of Discharges","Number of Readmissions"]].corr()
```
